```python
import numpy as np
import random
from scipy.ndimage import rotate, shift
import cv2

# Identify rows with class 0 (no tumor)
negative_indices = data[data.iloc[:, -1] == 0].index
print(f"Found {len(negative_indices)} negative class samples")

# Define how many more samples we need
target_new_samples = 60

# Define augmentation functions
def random_rotation(image, max_angle=20):
    angle = random.uniform(-max_angle, max_angle)
    return rotate(image, angle, reshape=False)

def random_shift(image, max_shift=20):
    x_shift = random.uniform(-max_shift, max_shift)
    y_shift = random.uniform(-max_shift, max_shift)
    return shift(image, [y_shift, x_shift])

def random_brightness(image, max_delta=0.2):
    delta = random.uniform(-max_delta, max_delta)
    image = image + delta
    return np.clip(image, 0, 1)  # Ensure values stay in valid range

def random_flip(image):
    if random.choice([True, False]):
        return np.fliplr(image)
    return image

# Function to apply multiple augmentations
def augment_image(image):
    # Reshape the flattened image back to 2D
    img_size = int(np.sqrt(len(image) - 1))  # -1 for the class label
    img = image[:-1].reshape(img_size, img_size)

    # Apply random augmentations
    img = random_rotation(img)
    img = random_shift(img)
    img = random_brightness(img)
    img = random_flip(img)

    # Flatten and append the class label
    augmented = list(img.flatten()) + [image[-1]]
    return augmented

# Create new augmented samples
new_samples = []
indices_to_augment = np.random.choice(negative_indices, min(target_new_samples, len(negative_indices)), replace=True)

for idx in indices_to_augment:
    original_sample = data.iloc[idx].values
    augmented_sample = augment_image(original_sample)
    new_samples.append(augmented_sample)

# Create DataFrame with new samples
augmented_df = pd.DataFrame(new_samples, columns=data.columns)

# Combine original and augmented data
combined_df = pd.concat([data, augmented_df], ignore_index=True)

print(f"Original DataFrame shape: {data.shape}")
print(f"Added {len(new_samples)} augmented negative class samples")
print(f"New DataFrame shape: {combined_df.shape}")

# Verify new class distribution
classes = combined_df.iloc[:, -1]
unique_classes = classes.unique()
class_counts = [sum(classes == cls) for cls in unique_classes]

plt.figure(figsize=(8, 5))
plt.bar(unique_classes, class_counts)
plt.title('Updated Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

# Actual CNN from scratch

parameters needed for conv layers:
- convolution is just applying a filter(kernel) over given input volume
so the filter or kernel has four main params:
1. width and height of kernel denoted by F
2. - in channels - channel coming from input
  - out channels - channels going into output
3. stride is the step denoted by S
4. padding is padding to image?


```python
class Convlayer2D:
  self __init__(kernel_size,in_channels,out_channels,stride,padding):
    # I think no of kernels = out channels
    '''
    kernel is just a square matrix of size kernel_size
    in_channels are channels in the input(depth of input)
    out_channels are channels required in output (depth of output ) which is same as no. of kernels
    stride is the step with which the kernel moves over
    padding is 0's padded on the input
    '''
    for i in range(out_channels):
      kernel = torch.randn()

```

```python
def conv2D(kernel,input_img,stride=1):
  '''
  This is the convolution function that'll slide the kernel over the input matrix
  kernel : filter matrix
  input : input matrix to convolve
  stride: step with which the kernel proceeds. default is 1
  '''
  kH, kW = len(kernel), len(kernel[0]) # height is number of rows in kernel, width is number of columns

  iH, iW = len(input_img), len(input_img[0]) # height is number of rows in input, width is number of columns in input

  # Formula for output width and height:
  output_W = (iW - kW) // stride + 1
  output_H = (iH - kH)// stride + 1

  output = [[0]* output_W for i in range(output_H)]


  for row in range(output_H):
    for col in range(output_W):
      sum = 0
      for ki in range(kH):
        for kj in range(kW):
          sum += input_img[row * stride + ki][col * stride + kj] * kernel[ki][kj]
      output[row][col] = sum

  return output

kernel = [ [1,1,1]
          ,[1,1,1],
           [1,1,1]]

input =[
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3],
     [1,2,3,1,2,3,1,2,3]
]

conv2D(kernel,input)
```
